```
from scapy.all import *
dataPath = "/content/drive/My Drive/Colab Notebooks/Final/data/LINUX/CAMONLY"
dataFile = "/CAMONLY_NORMAL_00001_20200725232935.pcapng"
```

```
eth_headers=['dst','src','type']
ip_headers=['version','ihl','tos','len','id','flags','frag','ttl','proto','chksum','src','dst']
udp_headers=['sport','dport','len','chksum']
raw_load=['load']

headers=[]
for h in eth_headers:
  headers.append("eth_"+h)
for h in ip_headers:
  headers.append("ip_"+h)
for h in udp_headers:
  headers.append("udp_"+h)
for h in raw_load:
  headers.append("load")
if "time" not in headers:
  headers = ["time"]+headers
headers
```

```
import numpy as np
def extract(pkt):
  time=pkt.time
  eth_dst=pkt.dst
  eth_src=pkt.src
  eth_type=pkt.type
  ip_version=pkt["IP"].version
  ip_ihl=pkt["IP"].ihl
  ip_tos=pkt["IP"].tos
  ip_len=pkt["IP"].len
  ip_id=pkt["IP"].id
  ip_flags=pkt["IP"].flags
  ip_frag=pkt["IP"].frag
  ip_ttl=pkt["IP"].ttl
  ip_proto=pkt["IP"].proto
  ip_chksum=pkt["IP"].chksum
  ip_src=pkt["IP"].src
  ip_dst=pkt["IP"].dst
  udp_sport=pkt["UDP"].sport
  udp_dport=pkt["UDP"].dport
  udp_len=pkt["UDP"].len
  udp_chksum=pkt["UDP"].chksum
  load=pkt["Raw"].load

  return [time,eth_dst,eth_src,eth_type,ip_version,ip_ihl,ip_tos,ip_len,ip_id,ip_flags,ip_frag,ip_ttl,ip_proto,ip_chksum,ip_src,ip_dst,udp_sport,udp_dport,udp_len,udp_chksum,load]
```

```
import numpy as np
import pandas as pd
import os
import csv

def UTConcat(file):
  f = open((dataPath+dataFile).split('.')[0]+'.csv','w+',newline ='')
  fcsv= csv.writer(f)
  fcsv.writerow(headers)
  try: # Pass all unreadable files
    cap = rdpcap(file)
  except:
    print(file, " unreadable, go for the next.\n")
    return

  for pkt in cap:
    try: # Accept only udp packets
      fcsv.writerow(extract(pkt))
    except:
      continue
  print(file, " completed.\n")

  f.close()
  print("output file closed.\n")


```

```
UTConcat(dataPath+"/tcp"+dataFile)
```
